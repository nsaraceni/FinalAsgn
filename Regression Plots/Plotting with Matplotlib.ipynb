## Line Plots 

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
# Now we will plot using .plot() and haiti

haiti.plot()
```

```python
# pandas typically populates the x-axis automatically with theindex values years
# and the y axis population
# label the graph with a title and x/y axis labels

haiti.index = haiti.index.map(int)
haiti.plot(kind = 'line')

plt.title('Immagration from Haiti')
plt.ylabel('Number of Immagrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.plot(kind ='line')

plt.title('Immagration from Haiti')
plt.ylabel('Number of Immagrants')
plt.xlabel('Years')

# we are annotating the 2010 earth quake. This shows a quick not on the graph
# syntax is: plt.text(x,y, label)

plt.text(2000,6000, '2010 Earth Quake',)
plt.show()

```

Since the x-axis (years) is type 'integer', we specified x as a year. The y axis (number of immigrants) is type 'integer', so we can just specify the value y = 6000.

If the years were stored as type 'string', we would need to specify x as the index position of the year. Eg 20th index is year 2000 since it is the 20th year with a base year of 1980.

```python
# Get the data set for China and India, and display the dataframe.

df_CI = df_can.loc[['China', 'India'], years]
df_CI
```

```python
# Plot graph. We will explicitly specify line plot by passing in kind parameter to plot().

df_CI.plot(kind = 'line')

```

```python
# Recall that pandas plots the indices on the x-axis and the columns as individual lines on the y-axis.
# Since df_CI is a dataframe with the country as the index and years as the columns, we must first transpose the dataframe using transpose() method to swap the row and columns.

df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind ='line')
```

```python
# Orrrrr....

df_CI.index = df_CI.index.map(int)
df_CI.plot(kind = 'line')

plt.title('Immagration from china and india')
plt.xlabel('Number of Immagrants')
plt.ylabel('years')

plt.show()

# this example you are labeling the graph 
```

Note: How come we didn't need to transpose Haiti's dataframe before plotting (like we did for df_CI)?

That's because haiti is a series as opposed to a dataframe, and has the years as its indices.
